```
X=norm_df.drop(44, axis=1) #caracteristica
Y=norm_df[44]    #class
```

```
# Tweak using Matplotlib
#plt.ylim(0, None)
#plt.xlim(0, None)
#sns.lmplot(x='t1', y='fcm1', data=df_new,fit_reg=False,hue='marcador')
```

```
f_classif=SelectKBest(score_func=f_classif,k=4) #k=4 seleciona as 4 melhores caracteristicas
```

```
fit=f_classif.fit(X,Y)
```

```
features=fit.transform(X)
```

```
print(features)
```

```
columns=fit.get_support(indices=True)
norm_df.iloc[:,columns]
```

```
#RFE
```

```
from sklearn.linear_model import LogisticRegression
```

```
model=LogisticRegression(max_iter=5000)
```

```
from sklearn.feature_selection import RFE
```

```
rfe=RFE(model,4)
```

```
fit= rfe.fit(X,Y)
```

```
print('Nmero de features: {}'.format(fit.n_features_))
```

```
columns=fit.get_support(indices=True)
```

```
norm_df.iloc[:,columns]
```

```
#Feature Importance
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
model=RandomForestClassifier(n_estimators=10)
```

```
model.fit(X,Y)
```

```
print(model.feature_importances_)
```

```
features_importances=pd.DataFrame(model.feature_importances_,index=X.columns,columns=['importance']).sort_values('importance', ascending=False)
```

```
features_importances
```

```
features_importances.plot(kind='bar')
```
